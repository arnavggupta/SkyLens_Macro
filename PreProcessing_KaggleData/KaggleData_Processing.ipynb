```python
import os
import shutil


os.makedirs("/tmp/.kaggle", exist_ok=True)

shutil.copy(
    "/Volumes/skylens_macro/team1/kaggledata/kaggle.json",
    "/tmp/.kaggle/kaggle.json"
)

os.chmod("/tmp/.kaggle/kaggle.json", 0o600)


os.environ["KAGGLE_CONFIG_DIR"] = "/tmp/.kaggle"

print("✅ Kaggle credentials configured!")
```

```python
import kaggle

download_path = "/tmp/flight-delays"
os.makedirs(download_path, exist_ok=True)

kaggle.api.authenticate()
kaggle.api.dataset_download_files(
    "usdot/flight-delays",
    path=download_path,
    unzip=True
)

print("✅ Dataset downloaded!")
print("Files downloaded:", os.listdir(download_path))
```

```python
target_volume_path = "/Volumes/skylens_macro/team1/kaggledata/raw/"
os.makedirs(target_volume_path, exist_ok=True)

for file_name in os.listdir(download_path):
    src = os.path.join(download_path, file_name)
    dst = os.path.join(target_volume_path, file_name)
    shutil.copy(src, dst)
    print(f"✅ Uploaded: {file_name}")

print("\n🎉 All files uploaded to Volume!")
```
